```python
iris = pd.read_csv('https://stepik.org/media/attachments/course/4852/iris.csv', index_col=0)
```

```python
iris.head()
```

```python
for column in iris:
    ax = sns.kdeplot(iris[column], legend=True)
ax.legend(iris)
# Petal length имеет наибольший размах значений
# Sepal width и sepal length имеют унимодальное распределение
# Petal width и petal length имеют бимодальное распределение
```

```python
sns.violinplot(iris, y='petal length')
```

```python
sns.pairplot(iris, hue='species')
# какая из пар на графике имеет наибольшую корреляцию? Чем ближе они к линии 45 градусов, тем выше корреляция. 
# petal width и petal length
```
