```python
greens = pd.read_csv('http://stepik.org/media/attachments/course/4852/algae.csv')
```

```python
greens.head()
```

```python
mean_concentrations = greens.groupby('genus').aggregate({'sucrose': 'mean', 
                                                         'alanin': 'mean',
                                                         'citrate': 'mean', 
                                                         'glucose': 'mean', 
                                                         'oleic_acid': 'mean'})
```

```python
mean_concentrations
```

```python
np.around(greens.query("genus == 'Fucus'").alanin.describe().loc[['min', 'mean', 'max']].values, decimals=2)
# чему равны минимальная, средняя и максимальная концентрации аланина (alanin) среди видов рода Fucus. 
# Округлите до 2-ого знака, десятичным разделителем является точка.
```

```python
greens.groupby('group').aggregate({'species': 'count'})
# Число видов бурых, зеленых и красных водорослей
```

```python
greens.groupby('group').aggregate({'citrate': 'var'})
# дисперсия концентрации цитрата в бурых, зеленых и красных водорослях
```

```python
greens.groupby('group').aggregate({'sucrose': lambda x: max(x) - min(x)})
# размах (максимум - минимум) сахарозы в бурых, зеленых и красных водорослях
```
